*   The sum of 5 and 5 is less than 10
*   The value of 10//2 is equal to 2 + 3


*   The sum of 1, 5, 2, 4 is greater than 15
*   157 is even

*   The sum of 5 squared and 12 squared is equal to 169









```python
#The sum of 5 and 5 is less than 10
a = 5+5
print(a < 10)

#The value of 10//2 is equal to 2 + 3
a = 10//2
b = 2+3
print(a == b)

#The sum of 1, 5, 2, 4 is greater than 15

a = 1+5+2+4
print(a >= 15)

#157 is even

a = 157
print(a%2 == 0)

#The sum of 5 squared and 12 squared is equal to 169

a = 5**2
b = 12**2
print(a+b == 169)
```
